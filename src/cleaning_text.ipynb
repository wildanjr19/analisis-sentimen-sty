```python
import pandas as pd
import re
import string
import emoji
```

```python
# cleaning function
def cleaning(text):
    
    # hapus urls
    text = re.sub(r'http\S+|www\.\S+', '', text)

    # Hapus mention
    text = re.sub(r'@\w+', '', text)

    # Hapus hashtag, hanya menyimpan kata setelah tanda #
    text = re.sub(r'#', '', text)

    # Hapus angka
    text = re.sub(r'\d+', '', text)

    # Hapus tanda baca
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Hapus emoji
    text = emoji.replace_emoji(text, replace='')

    # Hapus whitespace berlebih
    text = re.sub(r'\s+', ' ', text).strip()

    # Ubah menjadi huruf kecil
    text = text.lower()

    return text
```

```python
# import data
data = pd.read_csv('scrap_data.csv')
data.head(5)
```

```python
data.info()
```

```python
data = data.astype(str)
```

```python
# cleaning data
data['text'] = data['full_text'].apply(cleaning)
```

```python
# memilih beberapa kolom yang akan digunakan
data = data[['text']]
```
